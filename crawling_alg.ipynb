```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
```

```python
from selenium.common.exceptions import NoSuchElementException
import time
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
import pandas as pd
```

```python
driver=webdriver.Chrome(ChromeDriverManager().install())
```

```python
search_term='brazil+podcast+english'
url='https://www.youtube.com/results?search_query={}&sp=EgIQAg%253D%253D'.format(search_term)
```

```python
def bypass_popup(driver):
    try:
        consent_button_xpath = "/html/body/ytd-app/ytd-consent-bump-v2-lightbox/tp-yt-paper-dialog/div[4]/div[2]/div[5]/div[2]/ytd-button-renderer[2]/a/tp-yt-paper-button"
        time.sleep(4)
        consent = driver.find_element_by_xpath(consent_button_xpath)
        consent.click()
    except:
        print("no cookie popup")
```

```python
def search_and_get_channels(url):
    driver.get(url)
    bypass_popup(driver)
    
    channels= driver.find_elements_by_tag_name('ytd-channel-renderer')
    return channels
    
```

```python
def get_videos_from_channel(channels, index):
    time.sleep(2)
    channels[index].click()
    
    time.sleep(2)
    try:
        driver.find_element_by_xpath('/html/body/ytd-app/ytd-popup-container/tp-yt-iron-dropdown/div/yt-bubble-hint-renderer/div[2]/div[2]/yt-button-renderer/a/tp-yt-paper-button/yt-formatted-string').click()
    except:
        print('bubble hint no exist')
    
    time.sleep(2)
    driver.find_element_by_xpath('/html/body/ytd-app/div/ytd-page-manager/ytd-browse/div[3]/ytd-c4-tabbed-header-renderer/tp-yt-app-header-layout/div/tp-yt-app-header/div[2]/tp-yt-app-toolbar/div/div/tp-yt-paper-tabs/div/div/tp-yt-paper-tab[2]').click()
    
    time.sleep(2)
    videos = driver.find_elements_by_class_name('style-scope ytd-grid-video-renderer')
    
    time.sleep(2)
    video_list=[]
    for video in videos[:10]:
      title= video.find_element_by_xpath('.//*[@id="video-title"]').text
      views= video.find_element_by_xpath('.//*[@id="metadata-line"]/span[1]').text
      when= video.find_element_by_xpath('.//*[@id="metadata-line"]/span[2]').text
      url= video.find_element_by_xpath('.//*[@id="video-title"]').get_attribute('href')
      vid_item= {
          'title':title,
          'views':views,
          'posted':when,
          'url':url
      }
      video_list.append(vid_item)
    
    return video_list
```

```python
def mainprog(url_search_term, no_of_channels):
    video_list=[]
    for index in range(no_of_channels):
        channels=search_and_get_channels(url_search_term)
        video_listc=get_videos_from_channel(channels, index)
        video_list=video_list+video_listc
    
    dataset=pd.DataFrame(video_list)
    return dataset
```

```python
mainprog(url, 5)
```
